```python
from fastai.data.transforms import RandomSplitter

#Split the data in training and validation set (the alternative would be sklearn)
#In fastai this seems to return a list of indices
trn_split, val_split = RandomSplitter(seed=42)(df)

#the list of indices is now applied to the actual tensors (independent and dependent), so we get two tensors
trn_indep, val_indep = t_indep[trn_split], t_indep[val_split]
trn_dep, val_dep = t_dep[trn_split], t_dep[val_split]

#the default of fastai RandomSplitter seems to be 80/20 which can be seen when looking at the length
len(trn_indep), len(val_indep)
```

```python
#n_coeff is the number of features again (117 in our case)
n_coeff = t_indep.shape[1]
```

```python
import torch.nn.functional as F

#calculates the predictions of the model
def calc_preds(coeffs, indeps):
    #coefficients are in this tuple shape which is initialized in the code further below
    l1,l2,const = coeffs
    #@ refers to matrix multiplication, then apply ReLU
    res = F.relu(indeps@l1)
    #for the output layer also add a bias
    res = res@l2 + const
    #apply sigmoid in the end, so the result is between 0 and 1
    return torch.sigmoid(res)
```

```python
#calculates the loss which is the mean absolute error in our case
def calc_loss(coeffs, indeps, deps): return torch.abs(calc_preds(coeffs, indeps)-deps).mean()
```

```python
def one_epoch(coeffs, lr):
    #in an epoch we first calculate the loss
    loss = calc_loss(coeffs, trn_indep, trn_dep)
    #then we calculate the gradients
    loss.backward()
    #We update the coefficients with the function defined before
    with torch.no_grad(): update_coeffs(coeffs, lr)
    #we print out the current loss
    print(f"{loss:.3f}", end="; ")
```

```python
#for every feature we initialize a random weights and we do the same for the hidden layer
#we substract 0.5 in order to center them at 0

def init_coeffs(n_hidden=200):
    layer1 = (torch.rand(n_coeff, n_hidden)-0.5)/n_hidden
    layer2 = torch.rand(n_hidden, 1)-0.3
    #one bias for the output layer
    const = torch.rand(1)[0]
    #return as a tuple
    return layer1.requires_grad_(),layer2.requires_grad_(),const.requires_grad_()
```

```python
def train_model(epochs=50, lr=100):
    #don't touch the seed for random number
    #I did this and the model could not fit anymore
    #as mentioned in the video, when doing this manually one has to be cautious about parameter initialization
    torch.manual_seed(442)
    #first initialize coefficients
    coeffs = init_coeffs()
    #then performe backpropagation for the number of epochs
    for i in range(epochs): one_epoch(coeffs, lr=lr)
    return coeffs
```

```python
#Takes the coefficients at one point in time and updates them using the gradients and the learning rate
def update_coeffs(coeffs, lr):
    for layer in coeffs:
        #sub_ seems to be an internal function that directly subtracts something from the given tensor
        layer.sub_(layer.grad * lr)
        #resets the gradients
        layer.grad.zero_()
```

## Approach using a framework

Compare the approach from the fastAI book chapter 4 (also used in a similar way for the iris dataset in my learning portfolio)

```python
from fastai.tabular.all import *
set_seed(42)
```

```python
trn_indep = trn_indep.to(torch.float32)
trn_dep = trn_dep.to(torch.float32)
```

```python
#Create our model
torch.manual_seed(5)
simple_net = torch.nn.Sequential(
    torch.nn.Linear(117,20),
    torch.nn.ReLU(),
    torch.nn.Linear(20,1),
    #important! I did not do this in the beginning which caused problems because the predicted value was not always between 0 and 1
    torch.nn.Sigmoid()
)

#Uncomment for optional outputs
#print(list(simple_net.parameters()))
#initial predictions
#print('Initial predictions:')
#print(simple_net(x).flatten())

#learning rate
lr=0.05

#Create optimizer class
class BasicOptim:
    def __init__(self,params,lr): self.params,self.lr = list(params),lr

    def step(self, *args, **kwargs):
        for p in self.params: p.data -= p.grad.data * self.lr

    def zero_grad(self, *args, **kwargs):
        for p in self.params: p.grad = None

#Create instance of optimizer class
opt = BasicOptim(simple_net.parameters(), lr)

def calc_grad(xb, yb, model):
    preds = model(xb).flatten()
    loss = mse(preds, yb)
    global loss_value 
    loss_value = loss.item()
    loss.backward()

#Train
def train_epoch(model):
    calc_grad(trn_indep, trn_dep, model)
    opt.step()
    opt.zero_grad()

def train_model(model, epochs):
    for i in range(epochs):
        train_epoch(model)

train_model(simple_net, 3000)

#Output final loss
global loss_value
print(loss_value)
```

```python
#Performance metrics, but still on training data set

#if between 0 and 0.5 -> we predict 0
#if between 0.5 and 1 -> we predict 1

predictions = simple_net(trn_indep).flatten()
print('Predictions:')

print(predictions)
final_preds = torch.where(predictions < 0.5, 0, 1)

print('Final classification:')
print(final_preds)

print('True label y:')
print(trn_dep)

accuracy = torch.where(final_preds == trn_dep, 1., 0.).mean()
print('Accuracy:',accuracy)

#number of wrong predictions
wrongs = torch.where(final_preds == trn_dep, 0., 1.).sum()
print('Wrong instances:', wrongs)

#Accuracy is very high, but we are still on the training dataset
```
