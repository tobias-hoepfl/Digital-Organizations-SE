## Cleaning and preparation of data

```python
data[data['DayPassPriceAdult'] == 0]
```

Is it plausible that there are Resorts that are free? -> For now I leave them in the data set

There should be no skiing resort where the lowest point is higher than the highest point

```python
data[data['HighestPoint']<data['LowestPoint']]
```

Are there na-values?

```python
data.isna().sum()
#No, there are none
```

Remove index column 'Unnamed: 0' and 'Resort' because they clearly have no value for further analysis

```python
data = data.drop(['Unnamed: 0', 'Resort'], axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

#Encoding of country to be moved here, because otherwise there were be different countries in train and test data, which resulted in an error when predicting for test set
transformer = ColumnTransformer(
    transformers = [
      ('categorical', make_pipeline(OneHotEncoder(handle_unknown="ignore")), ['Country'])
      ], remainder='passthrough')

data_2 = transformer.fit_transform(data)
#back to df
data_3 = pd.DataFrame(data_2, columns=transformer.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data_3, test_size=0.2, random_state=42)
```

```python
#Seperate label from data
train_label = train_set[["remainder__DayPassPriceAdult"]].copy()
train_set = train_set.drop("remainder__DayPassPriceAdult", axis=1)

#do the same for test data
test_label = test_set[["remainder__DayPassPriceAdult"]].copy()
test_set = test_set.drop("remainder__DayPassPriceAdult", axis=1)
```

OneHotEncoder and Scaling on remaining columns

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

transformer = ColumnTransformer(
    transformers = [
      ('categorical', make_pipeline(OneHotEncoder(handle_unknown="ignore")), ['remainder__Snowparks', 'remainder__NightSki']),
      ('normal', make_pipeline(StandardScaler()), ['remainder__IntermediateSlope', 'remainder__HighestPoint', 'remainder__GondolaLifts', 'remainder__TotalSlope', 'remainder__DifficultSlope', 'remainder__SnowCannons', 'remainder__SurfaceLifts', 'remainder__LowestPoint'])
], remainder='passthrough')

transformed_data_train = transformer.fit_transform(train_set)
transformed_df_train = pd.DataFrame(transformed_data_train, columns=transformer.get_feature_names_out())
transformed_df_train.head()

#do same transformation on test set for evaluation
transformed_data_test = transformer.fit_transform(test_set)
transformed_df_test = pd.DataFrame(transformed_data_test, columns=transformer.get_feature_names_out())
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

#train the model on train data
lin_reg = LinearRegression()
lin_reg.fit(transformed_df_train, train_label)

#make preditions (still on training data) 
data_predictions_on_train_data = lin_reg.predict(transformed_df_train)
```

```python
data_predictions_on_train_data[:5]
```

```python
train_label.iloc[:5].values
```

Prediction looks good so far -> move on to the test data now

```python
#Move to the testing data

data_predictions_on_test_data = lin_reg.predict(transformed_df_test)
#does not work, because there are countries in the test set, not encountered in the train set
```

```python
data_predictions_on_test_data[:5]
```

```python
test_label.iloc[:5].values
```

Prediction looks not as good, but still ok -> Maybe there is an overfitting issue

Look at performance measures next:

```python
#Median of absolute error on training and test data
print(np.median(np.absolute(data_predictions_on_train_data-(train_label.to_numpy()))))
print(np.median(np.absolute(data_predictions_on_test_data-(test_label.to_numpy()))))

#There seems to be a minor overfitting issue, but the estimates are still pretty accurate 
#The model is only 9,5 euros of on average
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
print(mean_squared_error(train_label.to_numpy(), data_predictions_on_train_data, squared=False))
print(mean_squared_error(test_label.to_numpy(), data_predictions_on_test_data, squared=False))
#confirms results from MAE
```
